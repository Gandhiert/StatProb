## SOAL 1.
1. Diketahui tinggi badan siswa kelas A adalah 100,5 ; 99,7 ; 101 ; 101,3 ; 99,9 ; 98,9 ; 165 ; 102,2 ; 98,7 ; 102,9 ; dan 100. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

```python
# Membuat list dari tinggi badan siswa kelas A
tinggi_badan = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Menghitung mean, median, dan modus dari list
mean = sum(tinggi_badan) / len(tinggi_badan)
median = sorted(tinggi_badan)[len(tinggi_badan) // 2]
modus = max(tinggi_badan, key=tinggi_badan.count)

# Mencetak hasil perhitungan
print("Mean:", mean)
print("Median:", median)
print("Modus:", modus)
```

```python
# Mengimpor library Matplotlib
import matplotlib.pyplot as plt

# Membuat list dari tinggi badan siswa kelas A
tinggi_badan = [100.5, 99.7, 101, 101.3, 99.9, 98.9, 165, 102.2, 98.7, 102.9, 100]

# Menghitung mean, median, dan modus dari list
mean = sum(tinggi_badan) / len(tinggi_badan)
median = sorted(tinggi_badan)[len(tinggi_badan) // 2]
modus = max(tinggi_badan, key=tinggi_badan.count)

# Membuat histogram dengan bins=10 dan warna biru
plt.hist(tinggi_badan, bins=10, color="blue")

# Menambahkan garis vertikal yang menunjukkan mean, median, dan modus
plt.axvline(mean, color="red", label="Mean")
plt.axvline(median, color="green", label="Median")
plt.axvline(modus, color="yellow", label="Modus")

# Menambahkan judul, label sumbu, dan legenda
plt.title("Histogram Tinggi Badan Siswa Kelas A")
plt.xlabel("Tinggi Badan (cm)")
plt.ylabel("Frekuensi")
plt.legend()

# Menampilkan histogram
plt.show()
```

## SOAL 2.
2. Berikut adalah nilai UAS pada mata pelajaran matematika di kelas A: 95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75. Menggunakan Mean, Median, dan Modus, coba gambarkan pusat distribusi data berikut. Manakah yang paling bagus?

```python
# Membuat list nilai UAS
nilai = [95, 35, 50, 65, 70, 85, 25, 100, 80, 90, 45, 40, 35, 20, 75, 50, 40, 60, 80, 75]

# Menghitung mean (rata-rata)
total = 0
for n in nilai:
  total += n
mean = total / len(nilai)
print("Mean =", mean)

# Menghitung median (nilai tengah)
nilai.sort()
if len(nilai) % 2 == 0:
  # Jika jumlah data genap, median adalah rata-rata dari dua nilai tengah
  tengah = len(nilai) // 2
  median = (nilai[tengah - 1] + nilai[tengah]) / 2
else:
  # Jika jumlah data ganjil, median adalah nilai tengah
  median = nilai[len(nilai) // 2]
print("Median =", median)

# Menghitung modus (nilai yang paling sering muncul)
frekuensi = {}
for n in nilai:
  if n in frekuensi:
    frekuensi[n] += 1
  else:
    frekuensi[n] = 1
modus = None
maks = 0
for k, v in frekuensi.items():
  if v > maks:
    maks = v
    modus = k
print("Modus =", modus)
```

```python
# Mengimpor modul matplotlib
import matplotlib.pyplot as plt

# Membuat grafik batang
plt.bar(range(len(nilai)), nilai)
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.title("Grafik Batang Nilai UAS")

# Menambahkan garis vertikal untuk mean, median, dan modus
plt.axhline(mean, color="red", label="Mean")
plt.axhline(median, color="green", label="Median")
plt.axhline(modus, color="blue", label="Modus")

# Menampilkan legenda
plt.legend()

# Menampilkan grafik
plt.show()
```
