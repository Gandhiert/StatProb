## SOAL 1.
1. Dalam suatu liga sepak bola, dipilih secara acak 10 pemain dari klub yang berbeda dan diketahui jumlah gol yang mereka raih dalam musim ini adalah: 19, 10, 10, 12, 13, 15, 14, 13, 9, 17. Lakukan analisis terhadap sebaran data dari dataset yang diberikan. Manakah ukuran sebaran yang paling mampu menggambarkan sebaran data dengan baik?

```python
# Mengimpor modul numpy dan matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Menyimpan data jumlah gol dalam array numpy
gol = np.array([19, 10, 10, 12, 13, 15, 14, 13, 9, 17])

# Menghitung ukuran sebaran data seperti mean, median, mode, range, varians, dan standar deviasi
mean = np.mean(gol) # mean adalah rata-rata dari data
median = np.median(gol) # median adalah nilai tengah dari data yang telah diurutkan
mode = max(gol, key=lambda x: list(gol).count(x)) # mode adalah nilai yang paling sering muncul dalam data
range = np.max(gol) - np.min(gol) # range adalah selisih antara nilai maksimum dan minimum dalam data
var = np.var(gol) # varians adalah ukuran seberapa jauh data tersebar dari mean
std = np.std(gol) # standar deviasi adalah akar kuadrat dari varians

# Mencetak ukuran sebaran data
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Range:", range)
print("Varians:", var)
print("Standar deviasi:", std)

# Membuat histogram untuk menampilkan frekuensi data
plt.hist(gol, bins=5, edgecolor="black")
plt.xlabel("Jumlah gol")
plt.ylabel("Frekuensi")
plt.title("Histogram jumlah gol pemain sepak bola")
plt.show()
```

## SOAL 2.
2. Dalam tim olahraga, tinggi pemain adalah: 180 cm, 185 cm, 175 cm, 190 cm, 170 cm, 195 cm, 165 cm. Bagaimana sebaran data dari tinggi pemain ini?

```python
# Mengimpor modul statistics
import statistics

# Membuat daftar tinggi pemain
tinggi_pemain = [180, 185, 175, 190, 170, 195, 165]

# Menghitung nilai rata-rata (mean)
rata_rata = statistics.mean(tinggi_pemain)
print("Rata-rata tinggi pemain adalah", rata_rata, "cm")

# Menghitung nilai tengah (median)
tengah = statistics.median(tinggi_pemain)
print("Nilai tengah tinggi pemain adalah", tengah, "cm")

# Menghitung nilai modus (mode)
modus = statistics.mode(tinggi_pemain)
print("Nilai modus tinggi pemain adalah", modus, "cm")

# Menghitung nilai simpangan baku (standar deviasi)
simpangan_baku = statistics.stdev(tinggi_pemain)
print("Nilai simpangan baku tinggi pemain adalah", simpangan_baku, "cm")
```

```python
# Import library matplotlib.pyplot sebagai plt
import matplotlib.pyplot as plt

# Buat list tinggi pemain
tinggi = [180, 185, 175, 190, 170, 195, 165]

# Buat histogram dengan plt.hist()
plt.hist(tinggi)

# Beri judul dan label sumbu
plt.title("Histogram Tinggi Pemain Tim Olahraga")
plt.xlabel("Tinggi (cm)")
plt.ylabel("Frekuensi")

# Tampilkan histogram
plt.show()
```
